# Création du modèle

````python
class ConvNetMNIST(nn.Module):

    def __init__(self, image_input_size=28, nb_classes_to_predict=10, nb_activations_maps=20, kernel_size=5, pooling_kernel=2):
        """Create an object of the class. You can modify the parameters of the network through additional arguments."""
        # initialize the parent class
        super().__init__()

        # store the components of the network in attributes
        self.conv_1 = nn.Conv2d(
            in_channels=1,
            out_channels=nb_activations_maps//2,
            kernel_size=kernel_size
            ) # 28x28 -> 24x24(conv) -> 12x12 (pool)
        self.conv_2 = nn.Conv2d(
            in_channels=nb_activations_maps//2,
            out_channels=nb_activations_maps,
            kernel_size=kernel_size
            ) # 12x12 -> 8x8(conv) -> 4x4 (pool)
        self.relu = nn.ReLU()  # used 3 times
        self.pool = nn.AvgPool2d(kernel_size=pooling_kernel)  # used 2 times
        self.flatten = nn.Flatten()
        CNN_module_output_size = (
            ((
                ((image_input_size -  kernel_size + 1)//pooling_kernel) - kernel_size + 1)//pooling_kernel)
            )
        self.fc_1 = nn.Linear(
            CNN_module_output_size * CNN_module_output_size * nb_activations_maps, #wxhxc
            128)
        self.fc_2 = nn.Linear(128, nb_classes_to_predict)

        # Store the hook handle to be able to remove it later
        self.hook_handle_forward = self.conv_2.register_forward_hook(self.save_activations)

        # Store the hook handle to be able to remove it later
        self.hook_handle_backward = self.conv_2.register_full_backward_hook (self.save_gradient)

    def save_activations(self, module, input, output):
        """Saves the output activations of the hooked layer."""
        # Clear previous activations if any, to only keep the latest for a single forward pass
        self.activations = output.detach().cpu().numpy().squeeze()

    def save_gradient(self, module, grad_in, grad_out):
        """Saves the output activations of the hooked layer."""
        # Clear previous activations if any, to only keep the latest for a single forward pass
        self.guided_gradient =  grad_out[0].cpu().numpy().squeeze()

    def register_hooks(self):
        # Ensure no old hooks are active before registering new ones
        self.remove_hooks()
        self.hook_handle_forward = self.conv_2.register_forward_hook(self.save_activations)
        self.hook_handle_backward = self.conv_2.register_full_backward_hook(self.save_gradient)

    def remove_hooks(self):
        """Removes the registered forward hook."""
        if self.hook_handle_forward is not None:
            self.hook_handle_forward.remove()
            self.hook_handle_forward = None
        if self.hook_handle_backward is not None:
            self.hook_handle_backward.remove()
            self.hook_handle_backward = None

    def forward(self, x: torch.tensor) -> torch.tensor:
        """Forward pass through the neural network. Compute predictions."""
        pred = self.conv_1(x)
        pred = self.relu(pred)
        pred = self.pool(pred)
        pred = self.conv_2(pred)
        pred = self.relu(pred)
        pred = self.pool(pred)
        pred = self.flatten(pred)
        pred = self.fc_1(pred)
        pred = self.relu(pred)
        pred = self.fc_2(pred)

        return pred
````

Donc on travaille avec 576 pixels en sortie. Il y a eu donc 576 fois le filtre d'appliqué. Un filtre se compose des opérations suivantes:
*   Wieght * pixel + biais = résultat d'une case du filtre
*   Somme de l'ensemble des résultat du filtre

Donc il y a pour un pixel en sortie, $5×5$ fois l'opération $w_{i} \times x_{i} + 𝑏_{k}$ (biais partagé par filtre) et $5×5 - 1$ fois l'opération de somme.
L'opération $w_{i} \times x_{i} + 𝑏_{k}$ est elle-même composée de deux opérations.

Au final, pour un pixel d'output, on a 25 multiplication de poids, 25 addition de biais et 24 addition de résultat.

Donc pour 576 pixels de sortie, nous avons $$576 \times (25 + 25 + 24) = 42624$$ opérations.

Ces opérations sont répétés pour chaque filtre. Dans la première couche on a 10 filtres, donc on a au total pour la première convolution $$42624 \times 10 = 426240$$ opérations.

Si on sépare le nombre d'opération impliquée on a :
*   $576 \times 25 \times 10 =144000$ mutiplications
*   $576 \times (25 + 24) \times 10 = 282240$ additions

Une fois les 10 outputs obtenus, il restent à les passer dans la $𝑅𝑒𝐿𝑈$ qui n'est ni une addition ni une soustraction.

Ensuite, il y a une opération de MaxPooling.

A présent, nous allons définir les fonctions d'entraînement et de suivi des performances. 

````python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import clear_output
import torch
import torch.nn.functional as F
import numpy as np
from tqdm import tqdm

sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 15})


def plot_losses(train_losses, test_losses, train_accuracies, test_accuracies):
    """Function to plot training and validation losses and accuracies."""
    clear_output(wait=True)

    fig, axs = plt.subplots(1, 2, figsize=(13, 4))

    axs[0].plot(range(1, len(train_losses) + 1), train_losses, label='train')
    axs[0].plot(range(1, len(test_losses) + 1), test_losses, label='test')
    axs[0].set_ylabel('Loss')

    axs[1].plot(range(1, len(train_accuracies) + 1), train_accuracies, label='train')
    axs[1].plot(range(1, len(test_accuracies) + 1), test_accuracies, label='test')
    axs[1].set_ylabel('Accuracy')

    for ax in axs:
        ax.set_xlabel('Epoch')
        ax.legend()

    plt.show()


def train(model, optimizer, train_dataloader, val_dataloader, n_epochs=5, wandb_project=None):
    """Function for training a neural network.

    Parameters
    ----------
    model : nn.Module
        Neural network for training.
    optimizer : torch.optim.Optimizer
        Optimization algorithm for hyperparameter updates (e.g., SGD).
    train_dataloader : DataLoader
        DataLoader for the training dataset.
    val_dataloader : DataLoader
        DataLoader for the validation dataset.
    n_epochs : int, default 5
        Number of epochs for training the neural network.
    wandb_project : str, default None
        WandB project name for logging the training process (optional).
    """
    if wandb_project:
        import wandb
        wandb.init(project=wandb_project)
        wandb.watch(model)

    train_losses = []
    test_losses = []
    train_accuracies = []
    test_accuracies = []

    for epoch in range(n_epochs):
        model.train()
        batch_losses = []
        batch_accuracies = []

        for x_train, y_train in tqdm(train_dataloader, desc=f"Training Epoch {epoch + 1}/{n_epochs}"):
            y_pred = model(x_train)
            loss = F.cross_entropy(y_pred, y_train)

            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

            batch_losses.append(loss.item())
            batch_accuracies.extend((torch.argmax(y_pred, dim=-1) == y_train).cpu().numpy())

        train_losses.append(np.mean(batch_losses))
        train_accuracies.append(np.mean(batch_accuracies))

        model.eval()
        val_loss = []
        val_accuracy = []

        with torch.no_grad():
            for x_val, y_val in tqdm(val_dataloader, desc="Validation", leave=False):
                y_pred = model(x_val)
                loss = F.cross_entropy(y_pred, y_val)

                val_loss.append(loss.item())
                val_accuracy.extend((torch.argmax(y_pred, dim=-1) == y_val).cpu().numpy())

        test_losses.append(np.mean(val_loss))
        test_accuracies.append(np.mean(val_accuracy))

        if wandb_project:
            wandb.log(
                {
                    "train loss": train_losses[-1],
                    "train accuracy": train_accuracies[-1],
                    "val loss": test_losses[-1],
                    "val accuracy": test_accuracies[-1],
                }
            )

        plot_losses(train_losses, test_losses, train_accuracies, test_accuracies)
        print(
            f"Epoch {epoch + 1}/{n_epochs} | "
            f"Train Loss: {train_losses[-1]:.4f}, Train Accuracy: {train_accuracies[-1]:.4f} | "
            f"Val Loss: {test_losses[-1]:.4f}, Val Accuracy: {test_accuracies[-1]:.4f}"
        )
````

# Ajout du module GRAD-CAM pour expliquer les inférences

````python
def GRAD_CAM(model, image_tensor, true_label):
  #1e étape : Prédictions pour récupérer les valeurs d'activations de la dernière couche CNN (via hook designé dans le modèle directement)
  model.register_hooks()
  scores_predictions = model(image_tensor) # Les activtations sont calculés à ce moment-là, dans model.activations.

  #Trouver le score maximum et la classe associée
  score_predicted_class, predicted_class = torch.max(scores_predictions, dim=1)

  #2e étape Passage backward pour récupérer les gradients de la denrière couche CNN par rapport à la classe prédite
  score_predicted_class.backward() #Récupération des valeurs des gradients via le hook, récupérées dans model.guided_gradient

  #On remet à zéro les gradients pour être sûr de pas réentrainé le modèle par mégarde avec les nouveaux gradients calculés
  model.zero_grad()
  
  #3e étape : Calcul de la moyenne des gradients pour chaque carte d'activations
  gradients_aggregated = model.guided_gradient.mean(axis=(1,2))

  #4e étape : Faire la somme entre les activations et les gradients
  #On multiplie la moyenne des gradients d'une carte par les activations de cette même carte, et ce pour chaque carte
  weighted_activations = model.activations * gradients_aggregated[:, np.newaxis, np.newaxis]
  # On somme toutes les "cartes d'activations pondérées" entre elles, terme à terme, pour créer notre carte unique qui servira de heatmap
  sum_of_weighted_activations = np.sum(weighted_activations, axis = 0)
  # On passe le résultat dans une ReLU pour garder que les pixels qui contribuent positivement à la prédiction de la classe cible
  relu_sum_weighted_activations = np.maximum(sum_of_weighted_activations, 0)
  # La carte obtenue est de taille (8x8), donc il faut modifier sa taille pour qu'elle soit facilement superposable sur l'image initiale

  #6e étape : Upscale le heatmap obtenue pour qu'elle soit raccord avec l'image de départ
  upsampled_heatmap = cv2.resize(relu_sum_weighted_activations,
                               (image_tensor.size(3), image_tensor.size(2)),
                               interpolation=cv2.INTER_LANCZOS4)

  model.remove_hooks()

  #7e étape : Superposition heatmap et image de départ et visualisation
  fig, ax = plt.subplots(1, 2, figsize=(8, 8))

  # Input image
  ax[0].imshow(image_tensor[0].numpy().squeeze())
  ax[0].axis("off")
  ax[0].set_title(f"True Label: {true_label}") # Added true label


  # Overlay the heatmap
  ax[1].imshow(image_tensor[0].numpy().squeeze())
  ax[1].imshow(upsampled_heatmap, alpha=0.5, cmap='coolwarm')
  ax[1].axis("off")
  ax[1].set_title(f"Predicted Label: {predicted_class.item()}") # Added predicted label

  return relu_sum_weighted_activations, upsampled_heatmap
````

# Explication du fonctionnement
Dans le papier, les auteurs pensent que les CNN retiennent les infos spatiales, et que ces infos sont ensuites perdus dans les FCNN (qui servent de classifieur). Donc ils pensent que la dernière couche de CNN est celles avec le meilleur compromis entre la meilleur représentation spatiale et la sémantique haut-niveau (représentation d'object, de strcture) : "Furthermore, convolutional layers naturally retain spatial information which is lost in fully-connected layers, so we can expect the last convolutional layers to have the best compromise between high-level semantics and detailed spatial information.".

L'idée est donc de récupérer les zones calculées par le réseaux de neurones jusqu'au dernier CNN comme importante via l'output de la dernière couche du CNN. Cependant, cette importance est classe-agnostique, c'est à dire que le réseau de neurones a pu identifier différentes parties de l'image comme importantes, certaines zones importantes pour une classe A et d'autres importantes pour une classe B. Nous aimerions avoir une carte de chaleur qui nous montre les zones qui ont contribuées fortement à la prédiciton d'une classe choisie et uniquement cette classe. Pour ce faire, il faut ajouter une notion de gradient par rapport à la classe choisie. Nous allons prendre les gradients de la dernière couche, par rapport à la classe choisie pour ne garder que les poids qui contribuent positivement à la prédiction.  Puis, nous viendrons pondérer l'output récupéré du dernier CNN par ces gradients "de classe choisie" pour obtenir une carte de chaleur qui révèle les zones importantes pour la prédiction de cette classe.

# Description de l'implémentation
Un CNN a un output qui dépend de la taille de son input et du nombre de filtre qu'on souhaite créer (l'augmentation du nombre de channels au fil des CNN). Dans mon exemple, j'ai choisit d'avoir 20 couches de filtres. J'ai une sortie de CNN de taille 64(images) x 20(channels/nb filtres) x 8(width) x 8(height) (cf schéma du modèle plus haut). Ainsi, lorsqu'on va récupérer les sorties de la dernière couche de CNN, on va récupérer ce que les auteurs appellent des "Feature map activation" $A^{k}$ avec $k$ le nombre de channels. On en récupère une pour chaque channels, donc ici 20, de taille 8x8 (et bien sûr une pour chaque image).

Pour récupérer ces feature activation maps, on va devoir calculer l'inférence pour l'image étudiée, puis récupérer la sortie de cette couche. Pour récupérer facilement les sorties d'une couche, on peut utiliser des "hooks" en pytorch.
```
#1e étape :
model.register_hooks()
scores_predictions = model(image_tensor)
```

Ensuite, nous devons récupérer les gradients pour la dernière couche CNN, par rapport à la classe de notre choix (le plus souvent la vraie classe, ou alors la classe prédite). Ici, nous choisissons de prendre le gradient par rapport à la classe prédite.

"As shown in Fig. 2, in order to obtain the class-discriminative localization map Grad-CAM width u and height v for any class c, we first compute the gradient of the score for class c, $y^{c}$ (before the softmax), with respect to  feature map activations $A^{k}$ of a convolutional layer, i.e.$$\frac{\partial y^{c}}{\partial A^{k}_{ij}}$$

Ces gradients calculés par rapport à une classe sont appelés "guided_gradient".

```
#2e étape
#Trouver le score maximum et la classe associée
score_predicted_class, predicted_class = torch.max(scores_predictions, dim=1)
#Calcul des gradients par rapport à la classe prédite
score_predicted_class.backward()
```

Une fois les guided-gradient obtenus, nous allons faire ce que les auteurs appellent le "global average pooling", c'est à dire nous allons faire pour chaque feature map d'activation la moyennne de tout les gradients. Ainsi, nous allons nous retrouver avec $α^{k}$ gradients moyens, $k$ étant le nombre de channels/filtres que nous avons décidés au départ et qui sont produits par nos CNN. Ainsi, nous obtenons l'importance globale de chaque carte d'activation pour cette classe : "These gradients flowing back are global-average-pooled over the width and height dimensions (indexed by $i$ and $j$ respectively) to obtain the neuron importance weights $$α^{c}_{k} = \frac{1}{𝑍} ∑_{i}∑_{j}\frac{\partial y^{c}}{\partial A^{k}_{ij}}$$

```
 #3e étape
  gradients_aggregated = model.guided_gradient.mean(axis=(1,2))
```

Les auteurs proposent ensuite de pondérer les activations précédemment caculées par leur gradient moyens. En effet, on a $k$ cartes d'activations, et on a $k$ gradients moyens. Donc pour chaque $k$, on vient multiplier la carte par son gradient. Ainsi, on obtiens les cartes d'activations pondérées par leurs contributions moyennes pour la classe pour la carte en question : "We perform a weighted combination of forward activation maps" : $$α^{c}_{k}A^{k}$$

```
#4e étape : Faire le produit entre les activations et les gradients
weighted_activations = model.activations * gradients_aggregated[:, np.newaxis, np.newaxis]
```
Puis, nous faisons la somme de ces cartes pondérées pour obtenir la carte pondérée totale : $$\sum_{k}α^{c}_{k}A^{k}$$
```
sum_of_weighted_activations = np.sum(weighted_activations, axis = 0)
```

Cela nous permet d'obtenir une première carte de chaleur. Cependant, les valeurs de cette carte de chaleur peuvent être négatives, dans le cas où la zone entraîne une contribution négative pour la classe (et certainement positive pour une autre classe). Comme nous somme surtout intéressés par identifier ce qui a contribué positivement à la prédiction, nous appliquons une ReLU pour garder que les valeurs positives : "and follow it by a ReLU" : $$𝐿^{c}_{Grad-CAM} = ReLU(∑_{k}α^{k}_{k}A^{k})$$ "We apply a ReLU to the linear combination of maps because we are only interested in the features that have a positive influence on the class of interest, i.e. pixels whose intensity should be increased in order to increase $y^{c}$. Negative pixels are likely to belong to other categories in the image. As expected, without this ReLU, localization maps sometimes highlight more than just the desired class and perform worse at localization"

```
relu_sum_weighted_activations = np.maximum(sum_of_weighted_activations, 0)
```

Le résultat est une heatmap de la taille de l'output du dernier CNN, donc 8x8. Notre image de départ est de taille 28x28. Donc il n'est pas possible de simplement superposer les deux images. Il faut donc l'upuscale

```
#5e étape : Upscale le heatmap obtenue pour qu'elle soit raccord avec l'image de départ
  upsampled_heatmap = cv2.resize(relu_sum_weighted_activations,
                               (image_tensor.size(3), image_tensor.size(2)),
                               interpolation=cv2.INTER_LANCZOS4)
```

Puis simplement afficher les deux images supoersées l'une sur l'autre.
